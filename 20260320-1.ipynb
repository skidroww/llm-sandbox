```python
from langchain_core.runnables import RunnableBranch
from langchain_core.output_parsers import StrOutputParser
```

```python
math_prompt = PromptTemplate.from_template("당신은 수학 전문가입니다. 다음 질문에 대해 논리적으로 답하세요: {question}")
math_chain = math_prompt | llm | StrOutputParser()
physics_prompt = PromptTemplate.from_template("당신은 물리학 전문가입니다. 다음 질문에 대해 전문적인 용어를 사용하여 답하세요: {question}")
physics_chain = physics_prompt | llm | StrOutputParser()
```

```python
general_prompt = PromptTemplate.from_template("다음 질문에 친절하고 상세하게 답하세요: {question}")
general_chain = general_prompt | llm | StrOutputParser()
```

```python
classification_prompt = PromptTemplate.from_template(
    "주어진 질문이 수학(math), 물리학(physics), 일반(general) 중 어느 분야에 해당하는지 단어 하나로만 출력하세요.\n"
    "질문: {question}"
)
```

```python
classification_chain = classification_prompt | llm | StrOutputParser()
```

```python
branch = RunnableBranch(
    (lambda x: "math" in x["topic"].lower(), math_chain),
    (lambda x: "physics" in x["topic"].lower(), physics_chain),
    general_chain
)
full_chain = (
    {"topic": classification_chain, "question": lambda x: x["question"]}
    | branch
)
```

```python
response = full_chain.invoke({'question' : '특수상대성이론에 대해서'})
```

특수상대성이론(Special Theory of Relativity, STR)은 **아인슈타인**이 정립한 고전역학의 일반화로, 주로 다음 두 가정(공준)에서 출발합니다.

1) **상대성 원리**: 물리 법칙은 모든 관성계에서 동일하다.  
2) **빛의 속도 불변(등속광속) 원리**: 진공에서 빛의 속도 \(c\)는 모든 관성계에서 동일하게 측정된다(광원이 어떤 운동을 하든 관측자의 운동과 무관).

이 두 가정이 가져오는 핵심 결론들은 “시간과 공간”의 개념이 기존(뉴턴) 직관과 달라진다는 점입니다.

---

## 1. 로렌츠 변환과 시공간(민코프스키) 구조
특수상대성에서는 서로 다른 관성계 사이의 좌표 변환이 **갈릴레이 변환**이 아니라 **로렌츠 변환(Lorentz transformation)**을 따릅니다.  
예를 들어 한 축 방향 속도 \(v\)에 대해
\[
\gamma = \frac{1}{\sqrt{1-\frac{v^2}{c^2}}}
\]
\[
t'=\gamma\left(t-\frac{vx}{c^2}\right),\quad x'=\gamma(x-vt)
\]
처럼 시간과 공간이 섞이게 됩니다. 이를 바탕으로 시공간은 4차원 **민코프스키 시공간**으로 이해됩니다.

---

## 2. 시간 지연(Time Dilation)
움직이는 물체의 고유시간 \(\tau\)에 비해 관측자가 보는 시간 \(t\)는 증가합니다.
\[
\Delta t = \gamma \Delta \tau
\]
따라서 상대 속도가 클수록( \(v\to c\) ) **시간은 느리게 흐르는 것처럼** 측정됩니다.

---

## 3. 길이 수축(Length Contraction)
관측자 기준으로 움직이는 방향의 길이 \(L\)은 정지한 고유길이 \(L_0\)보다 짧아집니다.
\[
L = \frac{L_0}{\gamma}
\]
즉, **방향성 있는 수축(로렌츠 수축)**이 발생합니다.

---

## 4. 동시성의 상대성(Simultaneity is Relative)
뉴턴적 직관에서는 “어떤 기준계에서 동시에 일어난 사건은 모든 관성계에서 동시에 일어난다”는 생각이 암묵적이지만, 특수상대성에서는 그렇지 않습니다.  
서로 다른 관성계에서 사건들의 **동시성(동시 사건의 집합)**은 달라지며, 이는 로렌츠 변환에서 \(t'\)가 \(x\)에 의존하기 때문에 생깁니다.

---

## 5. 상대론적 속도 합성(Relativistic Velocity Addition)
단순 갈릴레이 덧셈 대신 다음 형태가 됩니다(1차원):
\[
u=\frac{u'+v}{1+\frac{u'v}{c^2}}
\]
이로부터 어떤 경우에도 물체의 속도는 \(c\)를 초과할 수 없습니다.

---

## 6. 운동량과 에너지의 상대론적 표현
고전역학에서 쓰던 운동량과 에너지 공식이 상대론적으로 수정됩니다.

- **상대론적 운동량**
\[
\mathbf{p}=\gamma m \mathbf{v}
\]
- **총 에너지**
\[
E=\gamma mc^2
\]
- **질량-에너지 등가**
\[
E^2=(pc)^2+(mc^2)^2
\]
특히 정지질량 \(m\)에 대해 정지 에너지가
\[
E_0=mc^2
\]
로 정의됩니다.

---

## 7. 통상 질량 증가 개념 대신(현대적 표현)
실무적으로는 “질량이 증가한다”는 표현보다, 위처럼 **운동량/에너지의 상대론적 정의**와 **시간 지연·로렌츠 계수**를 이용해 설명하는 방식이 일반적입니다. 핵심은 \(v\)가 커질수록 \(\gamma\)가 발산하므로, 유한한 힘으로 \(v=c\)에 도달하기 어렵다는 점입니다.
